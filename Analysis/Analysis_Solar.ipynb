```python
import pandas as pd
df = pd.read_parquet('/Users/darcor/Data/EDP Thesis/Output/Photovoltaic_rcp_8_5.parquet', engine="fastparquet")
```

```python
df['Celsius'] = df['t2m'] - 273.15
```

```python
df.dropna(subset=['Celsius'], inplace=True)
```

```python
df['Year'] = df['time'].dt.year
```

```python
df.groupby(["Year"]).agg(
    mean_temp=('Celsius', 'mean'),
    min_temp=('Celsius', 'min'),
    max_temp=('Celsius', 'max'),
    count=('Celsius', 'count')
).reset_index()
```
